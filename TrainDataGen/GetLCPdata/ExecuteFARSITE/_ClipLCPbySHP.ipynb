```python
lcpfilepath = '/home/qweddww123/farsite/TrainDataGen/GetLCPdata/LCP/lcp_extract'
shpfilepath = '/home/qweddww123/farsite/TrainDataGen/WildfireShapefiles'

for year in sorted(os.listdir(lcpfilepath), key=int):
    print(year, '***************************************************************************************************************')    
    for target_fire in sorted(os.listdir(lcpfilepath+'/'+year), key=str):
        path = lcpfilepath+'/'+year
        if(target_fire[0] == '.'):
            continue
        elif(target_fire.split('_')[0] == 'ak' or target_fire.split('_')[0] == 'AK'):
            continue
        elif(target_fire.split('_')[0] == 'hi'):
            continue
        else:
            shpdir = shpfilepath+'/'+year+'/'+target_fire
            print(target_fire)
            dirname = sorted(os.listdir(shpdir), key=str)[0]
            lcpfiledir = lcpfilepath+'/'+year+'/'+target_fire+'/'+target_fire+'.lcp'
            shppath = shpdir+'/'+dirname+'/'+"shptest.shp"
            try:
                shpfile = geopandas.read_file(shppath)
            except:
                continue
            x1 = str(shpfile.bounds.minx[0])
            y1 = str(shpfile.bounds.miny[0])
            x2 = str(shpfile.bounds.maxx[0])
            y2 = str(shpfile.bounds.maxy[0])

            print(x1, y2, x2, y1)

            cp_cmd = "cp "+lcpfilepath+'/'+year+'/'+target_fire+'/'+target_fire+".lcp"+" "+lcpfilepath+'/'+year+'/'+target_fire+'/'+"test1.lcp"

            os.system(cp_cmd)

            try:
                os.system("rm "+lcpfilepath+'/'+year+'/'+target_fire+'/'+"test1.lcp")
            except:
                continue
            cmd = "gdalwarp -s_srs ESRI:102039 "+lcpfilepath+'/'+year+'/'+target_fire+'/'+target_fire+".lcp"+" "+lcpfilepath+'/'+year+'/'+target_fire+'/'+"test1.lcp"
            os.system(cmd)

            cmd = "gdal_translate -projwin "+x1+" "+y2+" "+x2+" "+y1+" -of LCP "+lcpfilepath+'/'+year+'/'+target_fire+'/'+"test1.lcp"+" "+lcpfilepath+'/'+year+'/'+target_fire+'/'+"clipped.lcp"
            os.system(cmd)

            print()

print("Clipping LCP completed!!")
```
