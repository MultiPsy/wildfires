```python
import numpy as np
import pandas as pd
import rasterio
from osgeo import gdal
import os
```

```python
wildfire = '/home/qweddww123/farsite/TrainDataGen/WildfireShapefiles'
lcp = '/home/qweddww123/farsite/TrainDataGen/GetLCPdata/LCP/lcp_extract'

for year in sorted(os.listdir(wildfire), key=int):
    print(year, '***************************************************************************************************************')
    for target_fire in sorted(os.listdir(wildfire+'/'+year), key=str):
        path = wildfire+'/'+year
        if(target_fire[0] == '.'):
            continue
        elif(target_fire.split('_')[0] == 'ak' or target_fire.split('_')[0] == 'AK'):
            continue
        elif(target_fire.split('_')[0] == 'hi'):
            continue
        else:
            print(target_fire)
            shpdir = wildfire+'/'+year+'/'+target_fire
            wildfire_time = sorted(os.listdir(shpdir), key=str)[0]
            lcppath = lcp+'/'+year+'/'+target_fire
            try:
                for lcpfile in sorted(os.listdir(lcppath), key=str):
                    if lcpfile == "clipped.lcp":
                        lcpfilepath = lcppath+'/'+lcpfile
                        readfile = rasterio.open(lcpfilepath)
                        #elevation
                        elev_arr = np.array(readfile.read(1))
                        df = pd.DataFrame(elev_arr)
                        df.to_csv(shpdir+'/'+wildfire_time+'/'+"Elevation.csv", header=False, index=False)
                        #aspect
                        asp_arr = np.array(readfile.read(2))
                        df = pd.DataFrame(asp_arr)
                        df.to_csv(shpdir+'/'+wildfire_time+'/'+"Aspect.csv", header=False, index=False)
                        #slope
                        slo_arr = np.array(readfile.read(3))
                        df = pd.DataFrame(slo_arr)
                        df.to_csv(shpdir+'/'+wildfire_time+'/'+"Slope.csv", header=False, index=False)
                        #fuel models
                        fuel_arr = np.array(readfile.read(4))
                        df = pd.DataFrame(fuel_arr)
                        df.to_csv(shpdir+'/'+wildfire_time+'/'+"Fuel_models.csv", header=False, index=False)
                        #canopy cover
                        ca_cov = np.array(readfile.read(5))
                        df = pd.DataFrame(ca_cov)
                        df.to_csv(shpdir+'/'+wildfire_time+'/'+"Canopy_cover.csv", header=False, index=False)
                        #canopy height
                        ca_hgt = np.array(readfile.read(6))
                        df = pd.DataFrame(ca_hgt)
                        df.to_csv(shpdir+'/'+wildfire_time+'/'+"Canopy_height.csv", header=False, index=False)
                        #canopy base height
                        ca_bhgt = np.array(readfile.read(7))
                        df = pd.DataFrame(ca_bhgt)
                        df.to_csv(shpdir+'/'+wildfire_time+'/'+"Canopy_base_height.csv", header=False, index=False)
                        #canopy bulk density
                        ca_bden = np.array(readfile.read(8))
                        df = pd.DataFrame(ca_bden)
                        df.to_csv(shpdir+'/'+wildfire_time+'/'+"Canopy_bulk_density.csv", header=False, index=False)
                        #try:
                        #    os.system("rm "+shpdir+'/'+wildfire_time+'/'+"Canopy_bulk_height.csv")
                        #except:
                        #    continue
            except:
                continue
                    
print("Generate LCP semantics completed!!")
```
