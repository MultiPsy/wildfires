```python
import os
import numpy as np
import pandas as pd
import csv
```

```python
wildfire = '/home/qweddww123/farsite/TrainDataGen/WildfireShapefiles'
lcp = '/home/qweddww123/farsite/TrainDataGen/GetLCPdata/LCP/lcp_extract'

for year in sorted(os.listdir(wildfire), key=int):
    print(year, '***************************************************************************************************************')
    for target_fire in sorted(os.listdir(wildfire+'/'+year), key=str):
        path = wildfire+'/'+year        
        if(target_fire[0] == '.'):
            continue
        elif(target_fire.split('_')[0] == 'ak' or target_fire.split('_')[0] == 'AK'):
            continue
        elif(target_fire.split('_')[0] == 'hi'):
            continue
        else:
            print(target_fire)
            cnt = 0
            shpdir = wildfire+'/'+year+'/'+target_fire
            wildfire_time = sorted(os.listdir(shpdir), key=str)[0]
            print(shpdir+'/'+wildfire_time+'/'+"MLinput")
            try:
                for semantics in sorted(os.listdir(shpdir+'/'+wildfire_time+'/'+"MLinput"), key=str):
                    if semantics == "Elevation.csv":
                        Elevation = list(csv.reader(open(shpdir+'/'+wildfire_time+'/'+"MLinput/"+semantics)))
                        Elevation = np.array(Elevation)
                        #print("Elevation shape:", Elevation.shape)
                        cnt += 1
                    if semantics == "Aspect.csv":
                        Aspect = list(csv.reader(open(shpdir+'/'+wildfire_time+'/'+"MLinput/"+semantics)))
                        Aspect = np.array(Aspect)
                        #print("Aspect shape:", Aspect.shape)
                        cnt += 1
                    if semantics == "Slope.csv":
                        Slope = list(csv.reader(open(shpdir+'/'+wildfire_time+'/'+"MLinput/"+semantics)))
                        Slope = np.array(Slope)
                        #print("Slope shape:", Slope.shape)
                        cnt += 1
                    if semantics == "Fuel_models.csv":
                        Fuel_models = list(csv.reader(open(shpdir+'/'+wildfire_time+'/'+"MLinput/"+semantics)))
                        Fuel_models = np.array(Fuel_models)
                        #print("Fuel_models shape:", Fuel_models.shape)
                        cnt += 1
                    if semantics == "Canopy_cover.csv":
                        Canopy_cover = list(csv.reader(open(shpdir+'/'+wildfire_time+'/'+"MLinput/"+semantics)))
                        Canopy_cover = np.array(Canopy_cover)
                        #print("Canopy_cover shape:", Canopy_cover.shape)
                        cnt += 1
                    if semantics == "Canopy_height.csv":
                        Canopy_height = list(csv.reader(open(shpdir+'/'+wildfire_time+'/'+"MLinput/"+semantics)))
                        Canopy_height = np.array(Canopy_height)
                        #print("Canopy_height shape:", Canopy_height.shape)
                        cnt += 1
                    if semantics == "Canopy_base_height.csv":
                        Canopy_base_height = list(csv.reader(open(shpdir+'/'+wildfire_time+'/'+"MLinput/"+semantics)))
                        Canopy_base_height = np.array(Canopy_base_height)
                        #print("Canopy_base_height shape:", Canopy_base_height.shape)
                        cnt += 1
                    if semantics == "Canopy_bulk_density.csv":
                        Canopy_bulk_density = list(csv.reader(open(shpdir+'/'+wildfire_time+'/'+"MLinput/"+semantics)))
                        Canopy_bulk_density = np.array(Canopy_bulk_density)
                        #print("Canopy_bulk_density shape:", Canopy_bulk_density.shape)
                        cnt += 1
                    if semantics == "Temperature.csv":
                        Temperature = list(csv.reader(open(shpdir+'/'+wildfire_time+'/'+"MLinput/"+semantics)))
                        Temperature = np.array(Temperature)
                        #print("Temperature shape:", Temperature.shape)
                        cnt += 1
                    if semantics == "Humidity.csv":
                        Humidity = list(csv.reader(open(shpdir+'/'+wildfire_time+'/'+"MLinput/"+semantics)))
                        Humidity = np.array(Humidity)
                        #print("Humidity shape:", Humidity.shape)
                        cnt += 1
                    if semantics == "Percipitation.csv":
                        Percipitation = list(csv.reader(open(shpdir+'/'+wildfire_time+'/'+"MLinput/"+semantics)))
                        Percipitation = np.array(Percipitation)
                        #print("Percipitation shape:", Percipitation.shape)
                        cnt += 1
                    if semantics == "Wind_speed.csv":
                        Wind_speed = list(csv.reader(open(shpdir+'/'+wildfire_time+'/'+"MLinput/"+semantics)))
                        Wind_speed = np.array(Wind_speed)
                        #print("Wind_speed shape:", Wind_speed.shape)
                        cnt += 1
                    if semantics == "Wind_direction.csv":
                        Wind_direction = list(csv.reader(open(shpdir+'/'+wildfire_time+'/'+"MLinput/"+semantics)))
                        Wind_direction = np.array(Wind_direction)
                        #print("Wind_direction shape:", Wind_direction.shape)
                        cnt += 1
                    if semantics == "Cloud_cover.csv":
                        Cloud_cover = list(csv.reader(open(shpdir+'/'+wildfire_time+'/'+"MLinput/"+semantics)))
                        Cloud_cover = np.array(Cloud_cover)
                        #print("Cloud_cover shape:", Cloud_cover.shape)
                        cnt += 1
                    if semantics == "Number_of_points.csv":
                        Number_of_points = list(csv.reader(open(shpdir+'/'+wildfire_time+'/'+"MLinput/"+semantics)))
                        Number_of_points = np.array(Number_of_points)
                        #print("Number_of_points shape:", Number_of_points.shape)
                        cnt += 1

                input_img = np.stack([Elevation, Aspect, Slope, Fuel_models, Canopy_cover,
                                     Canopy_height, Canopy_base_height, Canopy_bulk_density,
                                     Temperature, Humidity, Percipitation, Wind_speed, Wind_direction, Cloud_cover,
                                     Number_of_points
                                     ])

                input_img = input_img.tolist()
                #write 3D array to csv file
                with open(shpdir+'/'+wildfire_time+'/'+"MLinput"+'/'+'input_img.csv', 'w', newline='') as csvfile:
                    writer = csv.writer(csvfile, delimiter=',')
                    writer.writerows(input_img)

            except:
                print()
                continue

            #read csv file and contain in 3D array
            
            with open(shpdir+'/'+wildfire_time+'/MLinput/'+'input_img.csv', 'r') as f:
                reader = csv.reader(f)
                examples = list(reader)

            nwexamples = []
            for row in examples:
                nwrow = []
                for r in row:
                    nwrow.append(eval(r))
                nwexamples.append(nwrow)
            nwexamples = np.array(nwexamples)
            print(nwexamples.shape)
            #df = pd.DataFrame(input_img)
            #df.to_csv(shpdir+'/'+wildfire_time+'/'+'input_img.csv', header=False, index=False)
            print()

print("Generating ML input files completed!!")
```
