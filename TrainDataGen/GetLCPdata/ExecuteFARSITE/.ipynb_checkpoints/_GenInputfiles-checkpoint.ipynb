```python
import os
from random import *
import datetime as dt
import pandas as pd
import numpy as np

RAWS_num = 24
shp_path = '/home/qweddww123/farsite/TrainDataGen/WildfireShapefiles'

def calander(months):
    if months == 4:
        return False
    if months == 6:
        return False
    if months == 9:
        return False
    if months == 11:
        return False
    else:
        return True

def calc_24(hours, RAWS_num):
    hours += RAWS_num * 100
    if hours % 100 == 0:
        hours = hours % 2400 - 41
        if hours < 0:
            return 2400 + hours
        else:
            return hours % 2400
    else:
        return hours % 2400 - 1

for years in sorted(os.listdir(shp_path), key=int):
    print(years, "**********************************************************************************")
    for wildfire_name in sorted(os.listdir(shp_path+'/'+years), key=str):
        if(wildfire_name[0] == '.'):
            continue
        elif(wildfire_name.split('_')[0] == 'ak' or wildfire_name.split('_')[0] == 'AK'):
            continue
        elif(wildfire_name.split('_')[0] == 'hi'):
            continue
        else:
            #print()
            year = ''
            months = ''
            days = ''
            hours = ''
            #read from shapefiles
            wildfire_time = sorted(os.listdir(shp_path+'/'+years+'/'+wildfire_name))[0]
            cnt = 0
            for i in wildfire_time:
                if i == '_':
                    cnt += 1
                    continue
                if cnt < 4:
                    year += i
                if cnt >= 4 and cnt < 6:
                    months += i
                if cnt >= 6 and cnt < 8:
                    days += i
                if cnt >= 9:
                    hours += i
                cnt += 1
            #save str wildfire_time to int
            try:
                int_year = int(year)
                int_months = int(months)
                int_days = int(days)
                int_start_hours = int(hours)
            except ValueError:
                continue
                
            print(wildfire_name)
            #print(wildfire_time)
            #print(int_year, int_months, int_days, int_start_hours)

            int_end_hours = calc_24(int_start_hours, RAWS_num-1)
            end_hours = str(int_end_hours)
            #print("end_hours =", end_hours)
            start_hours = str(int_start_hours)
            #print("start_time =", start_hours)

            int_burn_days = int_days
            int_burn_months = int_months
            RAWS_str = ""
            #wind_spd = str(randrange(2, 15))
            #wind_dir = str(randrange(0, 360))
            tem_arr = []
            hum_arr = []
            perci_arr = []
            wind_spd_arr = []
            wind_dir_arr = []
            cld_arr = []
            for i in range(RAWS_num):
                if int_start_hours >= 2400:
                    int_burn_days += 1
                    int_start_hours %= 2400
                if int_burn_days > 28 and int_burn_months == 2:
                    int_burn_days = 1
                    int_burn_months = 3
                if int_burn_days > 30 and calander(int_burn_months) != True:
                    int_burn_months += 1
                    int_burn_days = 1
                if int_burn_days > 31 and calander(int_burn_months) == True:
                    int_burn_months += 1
                    int_burn_days = 1
                if int_burn_months > 12:
                    int_burn_months = 1
                if int_days > 31:
                    int_burn_months += 1
                #year month day hour temperature humidity percipitation wind_speed wind_direction cloud_cover
                tem = str(randrange(50, 86))
                hum = str(randrange(10, 81))
                perci = str(0.00)
                wind_spd = str(randrange(2, 15))
                wind_dir = str(randrange(0, 360))
                cld = str(randrange(0, 66))
                str2 = str(int_year)+' '+str(int_burn_months)+' '+str(int_burn_days)+' '+str(int_start_hours)+' '+tem+' '+hum+' '+perci+' '+wind_spd+' '+wind_dir+' '+cld+'\n'
                tem_arr.append(tem)
                hum_arr.append(hum)
                perci_arr.append(perci)
                wind_spd_arr.append(wind_spd)
                wind_dir_arr.append(wind_dir)
                cld_arr.append(cld)
                RAWS_str += str2
                int_start_hours += 100

            #initialize arrays with first element of the arrays (600*600)
            tem_semantic = [[ tem_arr[0] for i in range(600) ] for i in range(600) ]
            hum_semantic = [[ hum_arr[0] for i in range(600) ] for i in range(600) ]
            perci_semantic = [[ perci_arr[0] for i in range(600) ] for i in range(600) ]
            wind_spd_semantic = [[ wind_spd_arr[0] for i in range(600) ] for i in range(600) ]
            wind_dir_semantic = [[ wind_dir_arr[0] for i in range(600) ] for i in range(600) ]
            cld_semantic = [[ cld_arr[0] for i in range(600) ] for i in range(600) ]

            filepath = shp_path+'/'+years+'/'+wildfire_name+'/'+wildfire_time
            print(filepath)
            #save the array as .csv files
            tem_nparr = np.array(tem_semantic)
            df = pd.DataFrame(tem_nparr)
            df.to_csv(filepath+'/'+'Temperature.csv', header=False, index=False)
            hum_nparr = np.array(hum_semantic)
            df = pd.DataFrame(hum_nparr)
            df.to_csv(filepath+'/'+'Humidity.csv', header=False, index=False)
            perci_nparr = np.array(perci_semantic)
            df = pd.DataFrame(perci_nparr)
            df.to_csv(filepath+'/'+'Percipitation.csv', header=False, index=False)
            wind_spd_nparr = np.array(wind_spd_semantic)
            df = pd.DataFrame(wind_spd_nparr)
            df.to_csv(filepath+'/'+'Wind_speed.csv', header=False, index=False)
            wind_Dir_nparr = np.array(wind_dir_semantic)
            df = pd.DataFrame(wind_Dir_nparr)
            df.to_csv(filepath+'/'+'Wind_direction.csv', header=False, index=False)
            cld_nparr = np.array(cld_semantic)
            df = pd.DataFrame(cld_nparr)
            df.to_csv(filepath+'/'+'Cloud_cover.csv', header=False, index=False)

            #print(RAWS_str)

            int_start_hours = int(hours)
            #if the value of the start hours is higher than the value of the end hours
            if int_start_hours > int_end_hours:
                cnt = 0
                if int_burn_days - int_days < 0:
                    if int_burn_months == 2:
                        burn_periods = (int_burn_days+28) - int_days - 1
                    else:
                        burn_periods = (int_burn_days+30) - int_days - 1
                else:
                    burn_periods = int_burn_days - int_days + 1
                burn_periods_str = ""
                int_burn_days = int_days
                int_burn_months = int_months
                int_start_hours2 = int_start_hours
                for i in range(burn_periods):
                    cnt += 1
                    if cnt == 1:
                        start_hours = str(int_start_hours)
                        end_hours = '2359'
                    elif cnt > 1 and cnt < burn_periods:
                        if int_end_hours-800 > 0:
                            start_hours = str(int_end_hours-800)
                            end_hours = str(int_end_hours)
                        else:
                            start_hours = '0'
                            end_hours = str(int_end_hours)
                    else:
                        if int_end_hours-800 > 0:
                            start_hours = str(int_end_hours-800)
                            end_hours = str(int_end_hours)
                        else:
                            start_hours = '0'
                            end_hours = str(int_end_hours)
                    #month day start_hour end_hour
                    if int_burn_days > 28 and int_burn_months == 2:
                        int_burn_days = 1
                        int_burn_months = 3
                    if int_burn_days > 30 and calander(int_burn_months) != True:
                        int_burn_months += 1
                        int_burn_days = 1
                    if int_burn_days > 31 and calander(int_burn_months) == True:
                        int_burn_months += 1
                        int_burn_days = 1
                    if int_burn_months > 12:
                        int_burn_months = 1
                        int_burn_days = 1
                    str1 = str(int_burn_months)+' '+str(int_burn_days)+' '+start_hours+' '+end_hours+'\n'
                    burn_periods_str += str1
                    int_burn_days += 1
                if (int_end_hours+1 ) % 60 == 0:
                    int_end_hours = int_end_hours + 1 + 40
                int_burn_days -= 1
                start_hours = str(int_start_hours2)
            else:
                cnt = 0
                if int_burn_days - int_days < 0:
                    if int_burn_months == 3:
                        burn_periods = (int_burn_days+28) - int_days + 1
                    else:
                        burn_periods = (int_burn_days+30) - int_days + 1
                else:
                    burn_periods = int_burn_days - int_days + 1
                burn_periods_str = ""
                int_burn_days = int_days
                int_burn_months = int_months                
                int_start_hours2 = int_start_hours
                for i in range(burn_periods):
                    cnt += 1
                    if cnt == 1:
                        start_hours = str(int_start_hours)
                        end_hours = str(int_end_hours)
                    elif cnt > 1 and cnt < burn_periods:
                        if int_end_hours-800 > 0:
                            start_hours = str(int_end_hours-800)
                            end_hours = str(int_end_hours)
                        else:
                            start_hours = '0'
                            end_hours = str(int_end_hours)
                    else:
                        if int_end_hours-800 > 0:
                            start_hours = str(int_end_hours-800)
                            end_hours = str(int_end_hours)
                        else:
                            start_hours = '0'
                            end_hours = str(int_end_hours)
                    #month day start_hour end_hour
                    if int_burn_days > 28 and int_burn_months == 2:
                        int_burn_days = 1
                        int_burn_months = 3
                    if int_burn_days > 30 and calander(int_burn_months) != True:
                        int_burn_months += 1
                        int_burn_days = 1
                    if int_burn_days > 31 and calander(int_burn_months) == True:
                        int_burn_months += 1
                        int_burn_days = 1
                    if int_burn_months > 12:
                        int_burn_months = 1
                        int_burn_days = 1
                    str1 = str(int_burn_months)+' '+str(int_burn_days)+' '+start_hours+' '+end_hours+'\n'
                    burn_periods_str += str1
                    int_burn_days += 1
                if (int_end_hours+1 ) % 60 == 0:
                    int_end_hours = int_end_hours + 2 + 40
                int_burn_days -= 1
                start_hours = str(int_start_hours2)

            #print(burn_periods_str)

            int_end_hours = calc_24(int_start_hours, RAWS_num-1)
            if (int_end_hours+1)%100 == 60:
                int_end_hours = int_end_hours + 1 + 40
                #print(int_end_hours)
            else:
                int_end_hours = int_end_hours + 1
                #print(int_end_hours, 11)

            input_string = "FARSITE INPUTS FILE VERSION 1.0\n\
FARSITE_START_TIME: "+months+' '+days+' '+start_hours+"\n\
FARSITE_END_TIME: "+str(int_burn_months)+' '+str(int_burn_days)+' '+str(int_end_hours)+"\n\
FARSITE_TIMESTEP: 60\n\
FARSITE_DISTANCE_RES: 30.0\n\
FARSITE_PERIMETER_RES: 60.0\n\
FARSITE_MIN_IGNITION_VERTEX_DISTANCE: 15.0\n\
FARSITE_SPOT_GRID_RESOLUTION: 15.0\n\
FARSITE_SPOT_PROBABILITY: 0.05\n\
FARSITE_SPOT_IGNITION_DELAY: 0\n\
FARSITE_MINIMUM_SPOT_DISTANCE: 30\n\
FARSITE_ACCELERATION_ON: 1\n\
FARSITE_FILL_BARRIERS: 1\n\
SPOTTING_SEED: 253114\n\
\n\
FARSITE_BURN_PERIODS: "+str(burn_periods)+"\n\
"+burn_periods_str+"\n\
\n\
FUEL_MOISTURES_DATA: 1\n\
0 3 4 6 70 100\n\
RAWS_ELEVATION: 4159\n\
RAWS_UNITS: English\n\
\n\
RAWS: "+str(RAWS_num)+"\n\
"+RAWS_str+"\n\
\n\
FOLIAR_MOISTURE_CONTENT: 100\n\
CROWN_FIRE_METHOD: Finney\n\
NUMBER_PROCESSORS: 1\n\
\n\
\n\
FLAMELENGTH:\n\
SPREADRATE:\n\
INTENSITY:\n\
CROWNSTATE:\n\
"

            #print(input_string)
            inputfilepath = shp_path+'/'+years+'/'+wildfire_name+'/'+wildfire_time+'/'+wildfire_time+".input"

            f = open(inputfilepath, "w")
            f.write(input_string)
            f.close()

            print()

print("input file generation completed!!")
```
